### Data Cleaning

```python
# Separating prices into 2 different columns
df[['price USD($)', 'price UAH(₴)']] = df['price'].str.split('•', expand = True)
df.head()
```

```python
# Checking for Null/NaN values
df.isna().sum()
```

#### Converting prices into INT

```python
df['price USD($)'] = df['price USD($)'].str.replace('$', '')
df['price USD($)'] = df['price USD($)'].str.replace(' ', '')
```

```python
df['price UAH(₴)'] = df['price UAH(₴)'].str.replace('грн', '')
df['price UAH(₴)'] = df['price UAH(₴)'].str.replace(' ', '')
df.head()
```

```python
df['price UAH(₴)'] = df['price UAH(₴)'].str.replace('зПДВ', '')
df['price UAH(₴)'] = df['price UAH(₴)'].str.strip()
```

```python
df['price USD($)'] = df['price USD($)'].astype(int)
df['price UAH(₴)'] = df['price UAH(₴)'].astype(int)
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Cleaning city column

# df['city'] = df['city'].str.replace('( від )', '')
df['city'] = df['city'].apply(lambda x: x.replace('()', ''))
df['city'] = df['city'].str.strip()
df.head(3)
```

```python
pd.set_option('display.max_colwidth', None)
```

#### Cleaning road_accident, has a lot of different sentences

```python
filtered_df = df[~df['road_accident'].str.contains('Не був у ДТП|Ми розпізнали')]
filtered_df.head(4)
```

```python
df['road_accident'] = df['road_accident'].apply(lambda x: 'Був в ДТП' if x != 'Не був у ДТП' else x)
df.head()
```

```python
# Deleting price column
df = df.drop(columns = 'price')
```

#### Cleaning and changing data type of mileage

```python
df['mileage'] = df['mileage'].apply(lambda x: x.replace('тис. км', '').strip())
df.head()
```

```python
df['mileage'] = df['mileage'].apply(lambda x: 0 if x == 'без пробігу' else x)
df.head()
```

```python
df['mileage'] = df['mileage'].astype(int)
df.dtypes
```

#### Adding 000 to our mileage, because it's thousands of km

```python
df['mileage'] = df['mileage'].apply(lambda x: x * 1000)
df.head(3)
```

#### Changing column names and reordering

```python
df.rename(columns = {
    'road_accident': 'accident'
}, inplace = True)
```

```python
df = df[['name', 'city', 'mileage', 'fuel', 'transmission', 'accident', 'price USD($)', 'price UAH(₴)', 'link']]
df.head()
```

```python
df.to_csv(r"C:\\Users\\user\\Downloads\auto_ria.csv")
```

### Adding Regions to our Dataset, because we only have city or village names

```python
df = pd.read_csv(r"C:\\Users\\user\\Downloads\auto_ria2.csv")
df.dtypes
```

```python
# Our Regions dataset, which we scraped from Wikipedia

regions = pd.read_csv(r'C:\Users\user\Downloads\city_region.csv')
regions = regions.drop(columns = 'Unnamed: 0')
regions.head()
```

```python
# Changing some values in regions dataset
def categorize_value(x):
    if x == 'Дніпро(Дніпропетровськ)': return 'Дніпро'
    
    elif x == 'Володимир-Волинський': return 'Володимир'
    elif x == 'БілаЦерква': return 'Біла Церква'
    elif x == 'КривийРіг': return 'Кривий Ріг'
    elif x == 'Кропивницький(Кіровоград)': return 'Кропивницький'
    elif x == 'ЖовтіВоди': return 'Жовті Води'
    
    elif x == 'Чорноморськ(Іллічівськ)': return 'Жовті Води'
    elif x == 'Благовіщенське(Ульяновка)': return 'Благовіщенське'
    elif x == 'Покровськ(Красноармійськ)': return 'Покровськ'
    
    elif x == 'Южне(Південне)': return 'Южне'
    elif x == 'Захарівка(Фрунзівка)': return 'Одеська область'
    elif x == 'Подільськ(Котовськ)': return 'Подільськ'
    
    elif x == 'Мирноград(Димитров)': return 'Мирноград'
    elif x == 'Слобожанське(Ювілейне)': return 'Дніпропетровська область'
    elif x == 'Лиман(КраснийЛиман)': return 'Лиман'
    elif x == 'Покров(Орджонікідзе)': return 'Покров'
    
    elif x == 'НоваОдеса': return 'Нова Одеса'
    elif x == 'КривеОзеро': return 'Миколаївська область'
    elif x == 'МурованіКурилівці': return 'Вінницька область'
    elif x == 'ЛиповаДолина': return 'Сумська область'
    elif x == 'ПетропавлівськаБорщагівка': return 'Київська область'
    elif x == 'ВеликаМихайлівка': return 'Одеська область'
    
    elif x == 'Пулини(Червоноармійськ)': return 'Житомирська область'
    elif x == 'ВеликийБурлук': return 'Харківська область'
    elif x == 'Торецьк(Дзержинськ)': return 'Торецьк'
    elif x == 'Бахмут(Артемівськ)': return 'Бахмут'
    
    elif x == 'НоваКаховка': return 'Нова Каховка'
    elif x == 'ЗалізнийПорт': return 'Херсонська область'
    elif x == 'ВеликаНовосілка': return 'Донецька область'
    elif x == 'Більмак(Куйбишеве)': return 'Запорізька область'
    
    elif x == 'Довжанськ(Свердловськ)': return 'Довжанськ'
    elif x == 'Вітовськийрайон(Жовтневийрайон)': return 'Миколаївська область'
    elif x == 'ВеликаЛепетиха': return 'Херсонська область'
    elif x == 'СтаницяЛуганська': return 'Луганська область'
    
    else:
        return x
```

```python
# Clean up auto_ria table
```

```python
df['city'] = df['city'].apply(lambda x: categorize_value(x))
df.head()
```

```python
# Renaming regions columns
regions.rename(columns = {
    'Місто': 'city',
    'Регіон': 'region'
}, inplace = True)
```

```python
# Merging regions table and auto_ria table to find regions that are matching
# with our cities and villages
df2 = df.merge(regions, how = 'right')
df2
```

```python
# Droping everything else, that are not in regions dataset
df2 = df2.dropna()
```

```python
# Cheking for Null/NaN values
df2.isna().sum()
```

```python
df2.shape
```

### Separating fuel

```python
df2['fuel'].value_counts()
```

```python
# If seller write information about fuel - we split fuel type and engine capacity
df2['fuel_type'] = df2['fuel'].apply(lambda x: x.split(',')[0].strip() if x.strip() != 'Не вказано' else 'Не вказано')
df2.head()
```

```python
df2.columns
```

```python
# Reordering
df2 = df2[['manufacturer', 'model_name', 'release_year', 'mileage', 'fuel_type', 'fuel','transmission', 'accident', 'price USD($)', 'price UAH(₴)', 'city', 'region', 'link']]
df2.head()
```

```python
df2['fuel_type'].apply(lambda x: x.replace('л.', '').strip()).unique()
```

```python
# Choosing only not numeric values, we dont need 1.6/2.99 etc. as a fuel type
df2['fuel_type'] = df2['fuel_type'].apply(lambda x: ''.join([c for c in x if not c.isnumeric()]))
df2['fuel_type'] = df2['fuel_type'].apply(lambda x: x.replace('. л.', 'Не вказано'))
df2['fuel_type'] = df2['fuel_type'].apply(lambda x: x.replace(' л.', 'Не вказано'))
```

```python
# Cleaning the rows, where there is no fuel_type
df2 = df2[df2['fuel_type'] != '']
```
